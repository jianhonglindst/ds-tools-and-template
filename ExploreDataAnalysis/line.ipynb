```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame(
    {
        'date': pd.date_range(start='2020-06-01', end='2020-07-31'),
        'value1': np.random.uniform(0, 10, 61),
        'value2': np.random.normal(300, 40, 61),
    }
)
data.head()
```

## Multiple Lines with Different Scale.

```python
# seaborn
sns.lineplot(data=data.value1, color="g")
ax2 = plt.twinx()
fig = sns.lineplot(data=data.value2, color="b", ax=ax2)

ax = data.plot(x="date", y="value1", legend=False)
ax2 = ax.twinx()
data.plot(x="date", y="value2", ax=ax2, legend=False, color="r")
ax.figure.legend()
```
