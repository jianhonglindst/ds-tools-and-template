## Packages

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
```

## Setting

```python
figsize = (16, 12)
```

## Fucntions

```python
# Read sklearn dataset and convert it to datafame
def sklearn_dataset_to_df(dataset):
    df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
    df['target'] = pd.Series(dataset.target)
    return df
```

## Read Data

```python
# Get Data: iris
iris = sklearn_dataset_to_df(dataset=load_iris())
iris.head()
```

## Data manipulation

```python
# Convert column names.
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris.head()
```

## Plot: matplotlib

```python
fig = plt.figure(1, figsize=figsize)
plt.scatter(iris.sepal_length, iris.sepal_width)
plt.title('IRIS')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.xlim(left=iris.sepal_length.min() - 0.5, right=iris.sepal_length.max() + 0.5)
plt.ylim(bottom=iris.sepal_width.min() - 0.5, top=iris.sepal_width.max() + 0.5)
```

```python
fig = plt.figure(1, figsize=figsize)
plt.scatter(iris.sepal_length, iris.sepal_width, c=iris.species, cmap=plt.cm.Set1, edgecolors='k')
plt.title('IRIS')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.xlim(left=iris.sepal_length.min() - 0.5, right=iris.sepal_length.max() + 0.5)
plt.ylim(bottom=iris.sepal_width.min() - 0.5, top=iris.sepal_width.max() + 0.5)
```

## Plot: seaborn

```python
fig, ax = plt.subplots(figsize=figsize)
# sns.despine(fig, left=True, bottom=True)
sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='sepal_width',
    size='petal_length',
    hue='species',
    ax=ax,
)
```

## Plot: pandas

```python
ax = iris.plot(
    x='sepal_length',
    y='sepal_width',
    kind='scatter',
    figsize=figsize,
    xlim=(iris.sepal_length.min() - 0.5, iris.sepal_length.max() + 0.5),
    ylim=(iris.sepal_width.min() - 0.5, iris.sepal_width.max() + 0.5)
)
```
